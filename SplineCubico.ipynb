```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as CubicSpline
```

```python
#Datos de ejemplo para interpolar (puedes cambiar estos valores)
x = np.array([1, 2, 3, 4, 5])
y= np.array([2, 3, 5, 7, 3])
```

```python
#Crear el spline c√∫bico
spline = CubicSpline.CubicSpline(x, y, bc_type='natural')

#Coeficientes del spline cubico
print("Coeficientes por tramo:\n", spline.c)
```

```python
#Graficar Spline
# Puntos para evaluar el spline (suave)
x_new = np.linspace(x.min(), x.max(), 100)
y_new = spline(x_new)

# Gr√°fica
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'ro', label='Datos originales')
plt.plot(x_new, y_new, 'b-', label='Spline c√∫bico')
plt.title('Interpolaci√≥n con Splines C√∫bicos', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
