```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def spline_lineal(x, y):
    n = len(x)
    
    for i in range(n-1):
        m = (y[i + 1] - y[i])/(x[i + 1] - x[i])
        b = y[i] - m*x[i]
        
        if(b > 0):
            print(m, 'x + ', b, sep = '', end = '')
        elif(b < 0):
            print(m, 'x + ', abs(b), sep = '', end = '')
        else:
            print(m, 'x', sep = '', end = '')

        print(" para x en [", x[i], ",", x[i + 1], "]", sep='')
    
    plt.plot(x, y, '*r', x, y)
    plt.show()
```

```python
#Asignar valores a x y a y
x = np.array([0, 1, 2, 3, 4])
y = np.array([2, 3, 5, 7, 3])

#Mandar llamar la función
spline_lineal(x, y)
```
